<h3> Convert from numeric 0/1 to factors</h3>
<ul>
  <li>anaemia</li>
  <li>diabetes</li>
  <li>high_blood_pressure</li>
  <li>sex</li>
  <li>smoking</li>
  <li>death_event</li>
</ul>
<h3> Relevel to No/Yes (if appropriate) </h3>

```python
### In R
%%R
## Convert `anaemia, diabetes, high_blood_pressure, sex, smoking, death_event`
## from numeric 0/1 to factors and relevel to No/Yes (Except for Sex)

library(dplyr)
library(forcats)

binary_cols <- c("anaemia", "diabetes", "high_blood_pressure",
                 "sex", "smoking", "DEATH_EVENT")

# Step 1 — Convert numeric 0/1 → factor with levels "0","1"
df_fac1 <- df_csv |>
  mutate(
    across(any_of(binary_cols), ~ factor(., levels = c(0, 1)))
  )

# Step 2 — Relevel + relabel to "No","Yes" (order: No → Yes)
df_fac2 <- df_fac1 |>
  mutate(
    across(all_of(setdiff(binary_cols, "sex")),
           ~ fct_recode(.,
                        "No"  = "0",
                        "Yes" = "1") |>
             fct_relevel("No", "Yes"))
  )

# Quick checks
str(df_fac2[intersect(binary_cols, names(df_fac2))])
```

```python
### In Python
# List of binary columns
binary_cols = ["anaemia", "diabetes", "high_blood_pressure",
               "sex", "smoking", "DEATH_EVENT"]

# Step 1 — Convert numeric 0/1 → categorical (factor-like)
df_fac1 = df_csv_py.copy()
for col in binary_cols:
    if col in df_fac1.columns:
        df_fac1[col] = df_fac1[col].astype("category")

# Step 2 — Relabel to "No" / "Yes" and order the categories
df_fac2 = df_fac1.copy()
# exclude 'sex'
cols_to_convert = [c for c in binary_cols if c.lower() != "sex" and c in df_fac2.columns]
for col in cols_to_convert:
    if col in df_fac2.columns:
        df_fac2[col] = (
            df_fac2[col]
            .map({0: "No", 1: "Yes"})           # relabel
            .astype(pd.CategoricalDtype(        # relevel (No → Yes)
                categories=["No", "Yes"], ordered=True))
        )

# ✅ Quick check
df_fac2.dtypes
```

<h2 align="center" style="color:#1a237e; font-size:1.9em;">
🧮 Creating a New Column by Combining Information from Existing Variables
</h2>

<p align="center" style="font-size:1.1em; color:#263238;">
We generate a new feature that aggregates relevant details from several existing columns  
to capture a more comprehensive measure where appropriate.
</p>

We picked four lab-related columns from your dataset that reflect blood chemistry or metabolic status:

* creatinine_phosphokinase

* platelets

* serum_creatinine

* serum_sodium

These come from different physiological systems (muscle enzyme, blood cells, kidney function, electrolytes), but together they give a broad biochemical picture.

```python
### In R
%%R
# Load required packages
library(dplyr)
library(scales)

# Columns to normalize and average
blood_cols <- c("creatinine_phosphokinase", "platelets",
                "serum_creatinine", "serum_sodium")

# Step 1 — Standardize (z-score) each column
df_scaled <- df_fac2 %>%
  mutate(across(all_of(blood_cols),
                ~ as.numeric(scale(.x)),  # mean=0, sd=1
                .names = "{.col}_z"))

# Step 2 — Compute the row-wise average of standardized values
df_scaled <- df_scaled %>%
  rowwise() %>%
  mutate(blood_index = mean(c_across(ends_with("_z")), na.rm = TRUE)) %>%
  ungroup()

# ✅ Quick check of results
df_scaled %>%
  select(ends_with("_z"),blood_index) %>%
  head(n=5)%>%
  print(width = Inf)
```

```python
### In Python
from sklearn.preprocessing import StandardScaler

# Columns representing blood composition / chemistry
blood_cols = ["creatinine_phosphokinase", "platelets",
              "serum_creatinine", "serum_sodium"]

# Create a scaler and standardize each variable (mean=0, std=1)
scaler = StandardScaler()
df_fac2_scaled = df_fac2.copy()
df_fac2_scaled[blood_cols] = scaler.fit_transform(df_fac2_scaled[blood_cols])

# Compute the average (row-wise) to form a composite index
df_fac2_scaled["blood_index"] = df_fac2_scaled[blood_cols].mean(axis=1)

# ✅ Quick check
cols = [c for c in blood_cols if c != "blood_index"] + ["blood_index"]
print(df_fac2_scaled[cols].head())
```

<h2 align="center" style="color:#1a237e; font-size:1.9em;">
📊 Transforming Numeric Variables into Categorical Groups
</h2>

<p align="center" style="font-size:1.1em; color:#263238;">
This helps simplify interpretation and highlights group-level patterns in the data.
</p>


We took numeric columns like:

* age

* ejection_fraction

and transformed them into categorical versions using threshold cutoffs that make sense clinically or intuitively.

```python
### In R
%%R
# Example numeric variables to transform
num_to_cat <- c("age", "ejection_fraction")

# Start from your processed df_fac2 or df_scaled
df_cat <- df_fac2 %>%
  mutate(
    # Age group
    age_group = cut(age,
                    breaks = c(0, 40, 60, 80, Inf),
                    labels = c("Young", "Middle", "Older", "Elder")),

    # Ejection fraction group (heart pump strength)
    ef_group = cut(ejection_fraction,
                   breaks = c(-Inf, 35, 45, 60, Inf),
                   labels = c("Severely Reduced", "Moderately Reduced",
                              "Normal", "Above Normal")),
  )

# ✅ Quick check
df_cat %>%
  select(age, age_group, ejection_fraction, ef_group) %>%
  head()
```

```python
### In Python
import numpy as np

df_cat = df_fac2.copy()

# Age group
df_cat["age_group"] = pd.cut(
    df_cat["age"],
    bins=[0, 40, 60, 80, np.inf],
    labels=["Young", "Middle", "Older", "Elder"]
)

# Ejection fraction group
df_cat["ef_group"] = pd.cut(
    df_cat["ejection_fraction"],
    bins=[-np.inf, 35, 45, 60, np.inf],
    labels=["Severely Reduced", "Moderately Reduced", "Normal", "Above Normal"]
)


# ✅ Quick check
print(df_cat[["age", "age_group", "ejection_fraction", "ef_group"
              ]].head())
```

<h2> Bivariate Visualization </h2>

Bivariate analysis explores the relationship between two variables at a time — examining how one variable changes in relation to another.
It helps us understand association, correlation, and group differences, revealing patterns that univariate plots can’t show.

We’ll explore:

* Numeric vs Numeric → scatterplots with trend lines
* Numeric vs Categorical → boxplot, overlapped density plot, violin plot, jitter plot
* Categorical vs Categorical → grouped and stacked bar charts

```python
%%R
# Scatter plot: Numeric vs Numeric
ggplot(df, aes(x = age, y = time)) +
  geom_point(alpha = 0.6, color = "blue") +
  geom_smooth(method = "lm", color = "firebrick", se = TRUE) +
  labs(title = "Relationship between Age and Follow-up Time",
       x = "Age (years)",
       y = "Follow-up Duration (days)") +
  theme_minimal()

# The red line goes downward, which means older patients usually have shorter follow-up times.
# This happens because older people tend to have higher health risks, so their observation period in the study is often shorter.

# The points are spread out, showing that it’s not true for everyone —
# but overall, the pattern is clear: as age increases, follow-up time decreases.
```

```python
# Boxplot- Categorical vs Numeric
%%R
ggplot(df, aes(x = DEATH_EVENT, y = age, fill = DEATH_EVENT)) +
  geom_boxplot(alpha = 0.6) +
  scale_fill_manual(values = c("No" = "blue", "Yes" = "red")) +
  labs(title = "Boxplot: Age by Death Event",
       x = "Death Event",
       y = "Age (years)") +
  theme_minimal()

# This boxplot compares the age distribution between patients who survived and those who died.
# The group labeled (Death Event = Yes) generally has a higher median age,
# meaning older patients are more likely to have died in the study.
```

```python
%%R
# Overlapped density: Categorical vs Numeric
ggplot(df, aes(x = age, fill = DEATH_EVENT)) +
  geom_density(alpha = 0.45, position = "identity", color = NA) +
  scale_fill_manual(values = c("No" = "blue", "Yes" = "red")) +
  labs(title = "Overlapped Density: Age by Death Event",
       x = "Age (years)", y = "Density", fill = "DEATH EVENT") +
  theme_minimal()

# Overlapped density plots show how distributions differ between groups —
# Here, older patients had higher death rates, while younger patients were more likely to survive.
```

```python
# violin plot --> best plotting

%%R
# Violin plot: Categorical vs Numeric
ggplot(df, aes(x = DEATH_EVENT, y = age, fill = DEATH_EVENT)) +
scale_fill_manual(values = c("No" = "blue", "Yes" = "red")) +
  geom_violin(trim = FALSE, alpha = 0.6) +
  geom_boxplot(width = 0.1, fill = "white", outlier.shape = NA) +
  labs(title = "Violin Plot: Age by Death Event",
       x = "Death Event",
       y = "Age (years)") +
  theme_minimal()

# This violin plot shows how age is distributed among patients who survived versus those who died.
# The shape of each violin represents the distribution of ages — wide areas mean many patients fall in that age range.
# The plot shows that patients who died are generally older than those who survived.
```

```python
%%R
# Jitterplot – Categorical vs Numeric
ggplot(df, aes(x = DEATH_EVENT, y = age, color = DEATH_EVENT)) +
  geom_jitter(width = 0.5, alpha = 0.6, size = 3) +
  scale_color_manual(values = c("No" = "blue", "Yes" = "red")) +
  labs(
    title = "Jitter Plot: Age by Death Event",
    x = "Death Event",
    y = "Age (years)"
  ) +
  theme_minimal()

# This jitter plot shows the age of each patient, separated by death event status.
# The slight horizontal jitter spreads overlapping points, making it easier to see how many patients share similar ages.
# The pattern shows that patients who died are generally older than those who survived.
```

```python
%%R
# Bar plot: Categorical vs Categorical
ggplot(df, aes(x = high_blood_pressure, fill = DEATH_EVENT)) +
  geom_bar(position = "dodge") +
  scale_fill_manual(values = c("No" = "blue", "Yes" = "red")) +
  labs(
    title = "Bar Plot: High Blood Pressure vs Death Event",
    x = "High Blood Pressure",
    y = "Number of Patients",
    fill = "Death Event"
  ) +
  theme_minimal()

# This bar plot compares high blood pressure with death events among patients.
# The dodge position places survival and death bars side by side, making it easy to compare.
# If the heights of the bars are similar across 2 categories,
# it suggests that high blood pressure status did not strongly affect death rates in this dataset.
# If the “Died” bar is higher among high blood pressure, it would suggest that they experienced more deaths.

```

```python
%%R
# stacked bar plot- Categorical vs Ctegorical
ggplot(df, aes(x = high_blood_pressure, fill = DEATH_EVENT)) +
  geom_bar(position = "fill") +
  scale_fill_manual(values = c("No" = "blue", "Yes" = "red")) +
  labs(title = "Proportion Plot: Death Event by High Blood Pressure",
       x = "High Blood Pressure",
       y = "Proportion of Patients",
       fill = "Death Event") +
  theme_minimal()

# This proportional stacked bar plot shows the share of deaths and survivals within each 2 groups.
# Each bar represents 100% of patients in that high blood pressure category.
# The colored segments show what fraction of that group survived or died.
# If both bars have similar proportions, high blood pressure did not make much difference in survival outcomes.
# If the “Died” portion is noticeably larger among high blood pressure, it suggests they had a higher death rate.
```

* Faceting → small multiples (separate panels)
* Additional dimensions → animation, 3-D, interactivity


```python
%%R
# Scatterplot With Color (3 Variables)
ggplot(df, aes(x = age, y = ejection_fraction, color = DEATH_EVENT)) +
  geom_point(alpha = 0.6, size = 3) +
  scale_color_manual(values = c("No" = "blue", "Yes" = "red")) +
  geom_smooth(method = "lm", se = FALSE, color = "black") +
  labs(title = "Age vs Ejection Fraction Colored by Death Event",
       x = "Age (years)", y = "Ejection Fraction (%)",
       color = "Death Event") +
  theme_minimal()

# Color adds a third variable (Death Event).
# We can now see that patients who died (red) often have lower ejection fractions and are older.
```

```python
# --- Step 1: Calculate proportions manually ---
prop_df = (
    df.groupby(['smoking', 'DEATH_EVENT'])
      .size()
      .reset_index(name='count')
)

# Normalize counts within each smoking group (so each sums to 1)
prop_df['proportion'] = prop_df.groupby('smoking')['count'].transform(lambda x: x / x.sum())

# --- Step 2: Pivot for stacking ---
pivot_df = prop_df.pivot(index='smoking', columns='DEATH_EVENT', values='proportion').fillna(0)

# --- Step 3: Plot stacked bar (100% height per smoking category) ---
pivot_df.plot(
    kind='bar',
    stacked=True,
    color=['skyblue', 'salmon'],
    figsize=(7, 5),
    edgecolor='black'
)

# --- Step 4: Customize labels, theme, and legend ---
plt.title('Proportion Plot: Death Event by Smoking Status', fontsize=14, color='#1a237e', weight='bold')
plt.xlabel('Smoking (No = Non-smoker, Yes = Smoker)', fontsize=12)
plt.ylabel('Proportion of Patients', fontsize=12)
plt.legend(title='Death Event (No = Survived, Yes = Died)', frameon=False)
plt.grid(axis='y', linestyle='--', alpha=0.6)
sns.despine()
plt.tight_layout()
plt.show()
```
